## 1. Counting birds

### <u> Counting birds part 1 </u>

Every year there is a day where people are encouraged to count the birds in their garden. Joseph tries to participate every year. Below you see the data of Joseph's observations. Let's make a nice plot of the data, also include the total amount of birds spotted.

```python
# Sample data
data = {
    'year': ['2020', '2021', '2022', '2023', '2024'],
    'Pigeons': [33, 52, 26, 40, 38],
    'Sparrows': [54, 48, 66, 59, 61],
    'others': [104, 98, 88, 92, 101]
}
```

### Inspecting the data

Load and inspect the data.

```python
# Load and inspect
df = pd.DataFrame(data)

df
```

### Plotting the pigeon count

Plot the number of counted pigeons over the year, following these specifications:

* Line graph
* Size of the figure: (10,6)
* Full blue line, data point marked with a circle, linewidth 2, marker size 8
* Provide the necessary labels in order to interpret the graph and show a legend
* Provide a meaningfull title
* Show a grid

```python
#Plotting the Pigeon count
# importlib.reload(plt)
# print(plt.figure)
plt.figure(figsize=(10, 6))

plt.plot(df["year"], df["Pigeons"], color='blue', marker='o', linewidth='2', markersize='8', label='Pigeons')

plt.title("Number of pigeons over the years")

plt.xlabel("Year")
plt.ylabel("Pigeons")

plt.legend()
plt.show()


```

### <u> Counting birds part 3 </u>

### Bar chart of total
Display the data  using a bar chart (total) and a stacked bar chart (pigeons, sparrows, others). Ensure the total is displayed on the right side and the stacked bar on the left.
Make it informative (and pretty).

```python
# Your code
fig, (left_plot, right_plot) = plt.subplots(1, 2)


bar_width = 0.55

left_plot.bar(df["year"], df["Sparrows"], bar_width, color='blue', label='Sparrows')
left_plot.bar(df["year"], df["Pigeons"], bar_width, bottom=df["Sparrows"], color='red', label='Pigeons') 
left_plot.bar(df["year"], df["others"], bar_width, bottom=df["Pigeons"]+df["Sparrows"], color="yellow", label="others")
left_plot.set_title("Number of birds")
# left_plot.set_xlabel("Year")


right_plot.bar(df["year"], df["Total"], bar_width, color="purple")
right_plot.set_title("Total number of birds over the years")
# right_plot.set_xlabel('Year')

fig.text(0.5, 0.04, "Year")

fig.legend(loc='upper left')


```

### <u> Counting birds part 4 </u>

### Grouped bar chart

Display the birds data using a grouped bar chart. Place the count of the pigeons and sparrows on the left of the x-tick, others and total on the right. Ensure the bars do not overlap.

```python
# Your code

plt.figure(figsize=(10, 6))

bar_width = 0.2

x = range(len(df["year"]))

pigeons = plt.bar([i - bar_width * 1.5 for i in x], df["Pigeons"], bar_width, color='red', label='Pigeons', align="center")
sparrows = plt.bar([i - bar_width * 0.5 for i in x], df["Sparrows"], bar_width, color='blue', label='Sparrows')
others = plt.bar([i + bar_width * 0.5 for i in x], df["others"], bar_width, color="yellow", label='Others', align='center')
total = plt.bar([i + bar_width * 1.5 for i in x], df["Total"], bar_width, color='green', label='Total', align="center")

plt.bar_label(pigeons)
plt.bar_label(sparrows)
plt.bar_label(others)
plt.bar_label(total)

plt.xticks(x, df["year"])
plt.title("Number of birds over the years")

plt.xlabel("Year")
plt.ylabel("Number of birds")
plt.legend()
plt.show()
```
